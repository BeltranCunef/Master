#  Datos abiertos

En los últimos años, muchos gobiernos e instituciones públicas y privadas se han encargado de construir los llamados portales de datos abiertos. Estos portales tienen como objetivo final el almacenamiento de datos y la compartición de éstos con terceros

#### 1

Diseña una función llamada `municipios` cuyo dato de entrada sea el contenido del fichero y la salida sea la lista de municipios contenidos en el fichero. ¿Cuantos municipios hay? 

```python
# Sol: 
def municipios(fichero):
    aux = [] #lista donde iré guardando los nombres
    nombre = '' #guardo el nombre para comparar y no meter repetidos
    i = 0
    while i < len(fichero['data']): #recorro el fichero
        nombre1 = fichero['data'][i]['municipio_nombre'] #guardo en una variable el nombre
        if nombre1 == nombre: #lo comparo, si está paso de largo
            i = i + 1
            pass
        else: #en caso contrario lo añado
            i = i + 1
            nombre = nombre1
            aux.append(nombre1)
    return aux    
```

```python
a = municipios(data)
a, len(a)
```

Diseña una función llamada `nacimientos` cuyo dato de entrada sea el contenido del fichero y la salida sea el número total de nacimientos registrados en el fichero. 

```python
# Sol: 
def nacimientos(fichero):
    i = 0
    j = 0
    while i < len(fichero['data']): #recorro el fichero
        j = j + fichero['data'][i]['numero_nacimientos'] #voy sumando los nacimientos
        i = i + 1
    return j
```

#### 3

Diseña una función llamada `nacimientos_mujeres` cuyo dato de entrada sea el contenido del fichero y la salida sea el número total de niñas nacidas registradas en el fichero. 

```python
# Sol: 
def nacimientos_mujeres(fichero):
    i = 0
    j = 0
    while i < len(fichero['data']): #recorro el fichero
        if fichero['data'][i]['sexo'] == 'Mujer': #compruebo que sea mujer
            j = j + fichero['data'][i]['numero_nacimientos'] #sumo los nacimientos
        i = i + 1
    return j
```

#### 4

Diseña una función llamada `rangos_edad` cuyo dato de entrada sea el contenido del fichero y la salida sea la lista de los distintos rangos de edad contenidos en el fichero. 

```python
# Sol: 
def rangos_edad(fichero):
    aux = [] #lista donde guardo los rangos
    i = 0
    while i < len(fichero['data']): #recorro el fichero
        rango = fichero['data'][i]['rango_edad_de_la_madre'] #guardo el rango de la madre
        if rango in aux: #pregunto a ver si está en la lista auxiliar donde s evan guardando los rangos, si está paso 
            i = i + 1
            pass
        else: #en caso contrario lo añado
            i = i + 1
            aux.append(rango)
    return aux  
```

```python
d = rangos_edad(data)
d
```

```python
# Sol :
def nacimientos_mujeres_15(fichero):
    i = 0
    j = 0
    while i < len(fichero['data']): #recorro el fichero
        if fichero['data'][i]['rango_edad_de_la_madre'] == 'Menores de 15 años': #compruebo que el rango de edad es menos de 15
            j = j + fichero['data'][i]['numero_nacimientos'] #sumo los nacimientos
        i = i + 1
    return j
```

Diseña una función llamada `nacimientos_por_localidad` cuyo dato de entrada sea el contenido del fichero y la salida sea un nuevo diccionario donde las claves sean la localidad y el valor sea el número total de nacimientos en dicha localidad.

```python
f, l, g = nacimientos_por_localidad(data)
f, l
```

```python
#Ahora con diccionarios y, ¡funciona bien!

def nacimientos_por_localidad(fichero):
    nacimientos = {}
    i = 0
    while i < len(fichero['data']):
        if fichero['data'][i]['municipio_nombre'] in nacimientos:
            nacimientos[fichero['data'][i]['municipio_nombre']] = nacimientos[fichero['data'][i]['municipio_nombre']] + fichero['data'][i]['numero_nacimientos']
            i = i + 1
        else:
            nacimientos.update({fichero['data'][i]['municipio_nombre']:fichero['data'][i]['numero_nacimientos']})
            i = i + 1
    return nacimientos
```

```python
f = nacimientos_por_localidad(data)
f
```
